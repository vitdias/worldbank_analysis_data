```python
### Create a function to fetch new indicators based on selected list and then merge them to the previous df created

def add_data(df_left, indicatorId, country_ID, date_beg, date_end, sourceID):
    dic_temp = wbdata.get_data(indicatorId, country=country_ID, date=(str(date_beg), str(date_end)), source=sourceID) # Individuals using the Internet (% of population)
    print("  > ", indicatorId, "Chamou dic_temp !")

    indicator_name = dic_temp[0]['indicator']['value']
    print('  >> INDICATOR NAME: ', indicator_name)

    df_temp = pd.DataFrame({
        'country': [item['country']['value'] for item in dic_temp],
        'date': [item['date'] for item in dic_temp],
        indicator_name: [item['value'] for item in dic_temp]
    })

    df_enriched = pd.merge(df_left, df_temp, on=['country', 'date'], how='left')

    return df_enriched


selected_indicators = [
    # 'IT.NET.USER.ZS',
    'SP.POP.TOTL',
    'NY.GDP.PCAP.CD', #  GDP per capita (current US$)
    'SE.TER.CUAT.BA.ZS', # Educational attainment, at least Bachelor's or equivalent, population 25+, total (%) (cumulative)
    'IT.NET.BNDW',
    'IT.NET.BNDW.PC',
    'IT.NET.USER',
    # 'IT.NET.USER.P2', # No need to use, is the same as 1st indicator
    'fin14a.t.d',
    'fin14a1.d',
    'fin14b.t.d',
    'Internet',
    'gwp2_n_3',
    'gwp2_n_4',
    'IT.NET.USEC.CD',
    ]

# Select specific sources for some indicators
for ind in selected_indicators:
    try:
        if ind == 'IT.NET.USER.P2':
            sourceID = 12
        elif ind == 'IT.NET.BNDW':
            sourceID = 57
        elif ind == 'IT.NET.BNDW.PC':
            sourceID = 57
        elif ind == 'IT.NET.USEC.CD':
            sourceID = 57
        elif ind == 'IT.NET.USER':
            sourceID = 57
        else:
            sourceID = None

        df_pre_final = add_data(df_pre_final, ind, south_american_countries_code, 1990, 2023, sourceID)
        print('INDICATOR: ', ind, ' >> done!')
    except Exception as e:
        print('INDICATOR: ',ind, ' >> ERROR: ', e)

df_pre_final
```

```python
### Trying to fix population indicator that had problem when fetching data

def fetch_population_data(ind, countries):
    df_final_pop = pd.DataFrame()
    for country_code in countries:
        dic_temp = wbdata.get_data(ind, country=country_code)
        if dic_temp:
            indicator_name = dic_temp[0]['indicator']['value']

            df_temp = pd.DataFrame({
                'country': [item['country']['value'] for item in dic_temp],
                'date': [item['date'] for item in dic_temp],
                indicator_name: [item['value'] for item in dic_temp]
            })

            df_temp = df_temp[df_temp['date'].astype(int) >= 1990]
            df_final_pop = pd.concat([df_final_pop, df_temp], ignore_index=True)
    return df_final_pop

# Usando a funÃ§Ã£o para obter os dados de populaÃ§Ã£o
df_population_data = fetch_population_data('SP.POP.TOTL', south_american_countries_code)

# df_population_data['Population (in Millions), total'] = df_population_data['Population, total'] / 1000000

df_population_data
```

```python
pd.set_option('display.max_rows', 50) 
# Downloaded CSV in WORLD BANK WEBSITE

df_gdppc_world = pd.read_csv(rf'{os.getcwd()}\Downloaded files\API_NY.GDP.PCAP.CD_DS2_en_csv_v2_2001204.csv')
df_tot_pop = pd.read_csv(rf'{os.getcwd()}\Downloaded files\API_SP.POP.TOTL_DS2_en_csv_v2_2001050.csv')
df_it_users = pd.read_csv(rf'{os.getcwd()}\Downloaded files\API_IT.NET.USER.ZS_DS2_en_csv_v2_2008380.csv')

def adjust_df_downloaded(df):
    ind_name = df['Indicator Name'][0]
    columns_of_interest = ['Country Name'] + [str(year) for year in range(1990, 2023)]
    df_filtered = df[columns_of_interest]

    # here I use mel to turn it into a dataframe
    df2 = df_filtered.melt(id_vars=['Country Name'], 
                                    var_name='Year', 
                                    value_name=ind_name)
    return df2

df_gdppc_world2 = adjust_df_downloaded(df_gdppc_world)
df_tot_pop2 = adjust_df_downloaded(df_tot_pop)
df_it_users2 = adjust_df_downloaded(df_it_users)
df_it_users2
```

```python
html_content = response.text

# Parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table in the HTML
table = soup.find('table')  # ajuste se necessÃ¡rio para encontrar a tabela correta

# Extract rows from the table
rows = table.find_all('tr')

# Get the data
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)

# Create a DataFrame
df_requested = pd.DataFrame(data)

# Define the column names
df_requested.columns = ['Country', 'Monthly Internet Cost (In USD)', 'Median Speed (Mbps)', 'Value (Cost per Mbps)']
df_requested = df_requested.iloc[1:]

df_requested['Monthly Internet Cost (In USD)'] = df_requested['Monthly Internet Cost (In USD)'].replace('[\$,]', '', regex=True)
df_requested['Value (Cost per Mbps)'] = df_requested['Value (Cost per Mbps)'].replace('[\$,]', '', regex=True)
# df_requested = df_requested.applymap(lambda x: x.replace('.', ',') if isinstance(x, str) else x)


df_requested
```
